# Confidence intervals

```python
import numpy as np
from scipy import stats
from IPython.display import display, Markdown
```

```python
M = [3.875, 3.3870, 2.8082, 2.7636, 4.4473, 3.70960, 4.7407, 3.1641, 3.1803, 6.3510]

reps = len(M)

print("Rep.\tAvg Time")
for i in range(0, len(M)):
    print("%d\t%g" % (i+1,M[i]), sep=' ', end='\n')
```

```python
alpha = 0.05
confidence = (1-alpha)*100

mean = np.mean(M)
var = np.var(M,ddof=1)
desv = np.sqrt(var)

tval = stats.t.ppf(1-(alpha/2), reps-1)
hval = tval*(desv/np.sqrt(reps))

display(Markdown(
   rf"""
${confidence}\%$ confidence

$\bar{{X}} = {mean}$

$S = {desv}$

$t_{{n-1,1-\frac{{\alpha}}{2}}} = t_{{{reps-1},{1-alpha/2}}} ={tval}$

$h = {hval}$

Intervalo de confianza : $C.I. = (\bar{{X}}-h , \bar{{X}} + h) = ({mean-hval},{mean+hval})$

"""))
```
